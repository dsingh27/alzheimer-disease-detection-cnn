```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/home/divya/work/msc_project/adn_data.csv')

#loaded the csv file for the dataset 
```

```python
# Get statitics of dataframe
data.describe()
```

```python
# Group the data by Group and Sex, and get the count for each group
group_counts = data.groupby(['Group', 'Sex']).size().reset_index(name='Count')

# Pivot the data to create a table with Group as the rows, Sex as the columns, and Count as the values
pivoted_data = group_counts.pivot(index='Group', columns='Sex', values='Count')

# Create a bar chart
ax = pivoted_data.plot(kind='bar', stacked=True, rot=0)

# Add a title and axis labels
ax.set_title('Number of Males and Females in Each Group')
ax.set_xlabel('Group')
ax.set_ylabel('Count')

# Show the chart
plt.show()
```
